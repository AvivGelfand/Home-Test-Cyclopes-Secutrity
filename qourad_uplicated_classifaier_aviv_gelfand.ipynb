# Duplicates

```python
import pandas as pd

# Create an empty data frame to store the duplicates
duplicates_df = pd.DataFrame(columns=['Column', 'Duplicates_Count'])

# Loop through each column in the dataframe
for column in df.columns:
    # Calculate the count of duplicates in the column
    duplicates_count = df[column].duplicated().sum()
    
    # Create a temporary dataframe with the column name and duplicates count
    temp_df = pd.DataFrame({'Column': [column], 'Duplicates_Count': [duplicates_count]})
    
    # Concatenate the temporary dataframe with the duplicates_df
    duplicates_df = pd.concat([duplicates_df, temp_df], ignore_index=True)

# Print the duplicates_df
display(duplicates_df.T)
```

# find_similar_rows


```python

def find_similar_rows(df, col1, col2, n):
    def compare_strings(s1, s2):
        # Split the strings into words
        words1 = set(s1.split())
        words2 = set(s2.split())
        
        # Find the difference in words
        diff = words1.symmetric_difference(words2)
        
        # Return True if the difference is exactly n words, False otherwise
        return len(diff) == n
    
    # Apply the comparison function to each row and filter the DataFrame
    mask = df.apply(lambda row: compare_strings(row[col1], row[col1]), axis=1)
    return df[mask]

# Example usage
# n = 1  # Number of words that differ
# similar_rows = find_similar_rows(df, n)
for n in range(1, 2):
# Find rows with a difference of 2 words
# n = 1
    matching_rows = find_similar_rows(df=df, col1='question1', col2='question2', n=n)
    display(matching_rows)
```
